```python
import pandas as pd
import requests
import time
import json
import re
import numpy as np
import folium as fs
import geopandas
import altair as alt
```

Reading

```python
df_price = pd.read_csv('region_price.csv', header=None, names=['region', 'price'])
df_coords = pd.read_csv('hdb_coords.csv', header=0, usecols=[0,1,2], names=['region', 'long', 'lat'])
```

```python
df_price['region'] = df_price['region'].str.upper()
df_coords['region'] = df_coords['region'].str.upper()
```

```python
df_price.head(5)
```

```python

df_coords.head(5)
```

```python
df_price_coords = pd.merge(df_price, df_coords, on='region', how='inner')
df_price_coords.head(5)
```

```python
location = geopandas.GeoDataFrame(data=df_price_coords, geometry=geopandas.points_from_xy(df_price_coords['long'], df_price_coords['lat']))
location
```

```python
mapping = fs.Map(location=[1.2,103.8])
fs.Choropleth(geo_data = "master_plan_boundaries.json",
            data = location,
            columns = ['region','price'],
            key_on = 'feature.properties.PLN_AREA_C',#pathway to locate each town and its geometry in the geojson file
            fill_color = 'YlGnBu', # colour of choropleth
            nan_fill_color = "White", # color for empty data
            fill_opacity = 1, #density of colour
            line_opacity = 0.7,
            legend_name = 'Average Price Of Houses',
            line_color = 'black',
            nan_fill_opacity = 0.3).add_to(mapping)
```

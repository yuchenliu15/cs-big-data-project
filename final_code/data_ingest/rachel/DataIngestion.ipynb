```python
for year in years:
    for area in planning_areas:
        keys = "&planningArea={area}&year={year}".format(area = area, year = year)
        dict_area_stats['year'].append(year)
        dict_area_stats['planning_area'].append(area)
        for call in calls:
            url_wo_keys = "https://developers.onemap.sg/privateapi/popapi/{call}?token={token}".format(call = call, token = token)
            url = url_wo_keys + keys
            r = requests.get(url, verify = False)
            result = r.json()
            #if there is no data
            if('Result' in result):
                if(call == "getReligion"):
                    for key in religions:
                        dict_area_stats[key].append(0)
                if(call == "getSpokenAtHome"):
                    for key in languages:
                        dict_area_stats[key].append(0)
                if(call == "getEthnicGroup"):
                    for key in ethnicities:
                        dict_area_stats[key].append(0)
            else:
                data = result[0]
                for key in data:
                    if((key != "planning_area") and (key != "year") and (key != "others")):
                        value = data[key]
                        dict_area_stats[key].append(value)
    time.sleep(60)
                
                
```
